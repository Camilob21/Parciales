```python
%%writefile 1_üìà_Punto_1.py

import streamlit as st
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
st.set_page_config(page_title="Espectros", page_icon="üìà")

st.markdown("# Solucion Punto 1")
st.sidebar.header("Punto 1")
st.write(
    """Se analiza una funcion dada utilizando la serie de fourier"""
)

# Configuraci√≥n de Matplotlib
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Par√°metros de la se√±al
To = np.pi            # Periodo definido
Fo = 1/To
A = 1                 # Amplitud de la se√±al
F0 = 1/To           # Frecuencia fundamental
Fs = Fo*100         # Frecuencia de muestreo

# Intervalo de tiempo basado en F0
tv = np.arange(-To/2, To/2, 1/Fs)
Nm = len(tv)
# Se√±al original x(t)
xo = (A**2)/2 + (A**2 * np.cos(4 * np.pi * F0 * tv))/2

# --- Figura con subplots del an√°lisis ---
fig, axes = plt.subplots(4, 2, figsize=(14, 10))

# 1. Gr√°fica de x(t)
axes[0, 0].plot(tv, xo, 'r', linewidth=4)
axes[0, 0].grid()
axes[0, 0].set_xlabel("t[s]")
axes[0, 0].set_ylabel("x(t)")
axes[0, 0].set_title("Gr√°fica de x(t)")
axes[0, 0].set_xlim([-1/(2*F0), 1/(2*F0)])

# Calcular el espectro de Fourier
N = 50              # N√∫mero de arm√≥nicos a considerar
wo = 2*np.pi/To     # Frecuencia angular fundamental

# Definir las bases de Fourier
phin = np.zeros((Nm, 2*N+1), dtype=np.complex_)
for n in range(-N, N+1):
    phin[:, n+N] = np.exp(1j*n*wo*tv)

# Definir coeficientes (Cn) de la se√±al
nv = np.linspace(-N, N, 2*N+1)
Cn = np.zeros(2*N+1, dtype=np.float_)
Cn[N]   = (A**2)/2   # Componente DC (a0)
Cn[N+2] = (A**2)/4   # Componente correspondiente a cos(4*pi*F0*t)
Cn[N-2] = (A**2)/4   # Componente correspondiente a cos(-4*pi*F0*t)

# 2. Espectro: Parte real
axes[0, 1].stem(nv, np.real(Cn), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[0, 1].set_xlabel(r'$n$')
axes[0, 1].set_ylabel(r'$Re\{C_n\}$')
axes[0, 1].grid()
axes[0, 1].set_title("Parte Real del Espectro")
axes[0, 1].set_xlim(-5, 5)

# 3. Espectro: Parte imaginaria
axes[1, 0].stem(nv, np.imag(Cn), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[1, 0].set_xlabel(r'$n$')
axes[1, 0].set_ylabel(r'$Im\{C_n\}$')
axes[1, 0].grid()
axes[1, 0].set_title("Parte Imaginaria del Espectro")
axes[1, 0].set_xlim(-10, 10)

# 4. Espectro: Magnitud
axes[1, 1].stem(nv, np.abs(Cn), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[1, 1].set_xlabel(r'$n$')
axes[1, 1].set_ylabel(r'$|C_n|$')
axes[1, 1].grid()
axes[1, 1].set_title("Espectro de Magnitud")
axes[1, 1].set_xlim(-5, 5)

# 5. Espectro: Fase (radianes)
axes[2, 0].stem(nv, np.angle(Cn), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[2, 0].set_xlabel(r'$n$')
axes[2, 0].set_ylabel(r'$\angle C_n$')
axes[2, 0].grid()
axes[2, 0].set_title("Espectro en Fase")
axes[2, 0].set_xlim(-20, 20)

# 6. Espectro: Magnitud en dB
# Se a√±ade un peque√±o epsilon para evitar log10(0)
axes[2, 1].stem(nv, 20 * np.log10(np.abs(Cn) + 1e-10), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[2, 1].set_xlabel(r'$n$')
axes[2, 1].set_ylabel(r'$|C_n|$ [dB]')
axes[2, 1].grid()
axes[2, 1].set_title("Espectro |Cn| [dB]")
axes[2, 1].set_xlim(-20, 20)

# 7. Espectro: Fase en grados
axes[3, 0].stem(nv, np.angle(Cn, deg=True), linefmt='r-', markerfmt='ro', basefmt=" ")
axes[3, 0].set_xlabel(r'$n$')
axes[3, 0].set_ylabel(r'Fase [¬∞]')
axes[3, 0].grid()
axes[3, 0].set_title("Fase")
axes[3, 0].set_xlim(-20, 20)

# Se desactiva el subplot vac√≠o (8¬∫)
axes[3, 1].axis('off')

plt.tight_layout()
st.pyplot(fig)

# --- Secci√≥n interactiva: Se√±al reconstruida ---
st.markdown("## Se√±al reconstruida")

# Usamos un slider para seleccionar el par√°metro Na
Na = st.slider("Selecciona el valor de Na", min_value=1, max_value=N, value=1, step=1)

# Calcular los √≠ndices de los coeficientes a utilizar
ind = np.arange(N - Na, N + Na + 1)

# Reconstruir la se√±al utilizando los coeficientes seleccionados
xe = phin[:, ind].dot(Cn[ind])
Px = 3/8  # Energ√≠a total de la se√±al (seg√∫n la definici√≥n de Cn)
er = 1 - np.sum(np.abs(Cn[ind])**2) / Px  # Error de reconstrucci√≥n

# Crear figura para la se√±al reconstruida
fig2, ax2 = plt.subplots()
ax2.plot(tv, xo, 'r', label='Se√±al original x(t)')
ax2.plot(tv, xe, 'b', marker='o', markersize=3, label='Se√±al reconstruida $x_e(t)$')
ax2.set_title('$E_r$ = %.2f [%%]' % (100 * er), fontsize=16)
ax2.set_xlabel('t [s]')
ax2.set_ylabel('x(t)')
ax2.grid()
ax2.legend()

st.pyplot(fig2)
```

```python
%%writefile 2_üìª_Se√±al_Modulada.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import yt_dlp as youtube_dl
import subprocess
import os
import io

# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="Se√±al Modulada", page_icon="üìª")

st.markdown("# Se√±al Modulada")
st.sidebar.header("Punto 2")
st.write(
    """Esta demo muestra la modulaci√≥n de una se√±al a partir de una canci√≥n descargada de YouTube.
Se descarga la canci√≥n, se extrae una muestra, se modula y se visualizan las se√±ales en tiempo y frecuencia.
Adem√°s, se reproducen las se√±ales resultantes."""
)

# --- Funci√≥n para descargar el audio de YouTube ---
def descarga(video_url, name):
    try:
        # Se simplifica la descarga pasando directamente el URL
        ydl_opts = {
            'format': 'bestaudio/best',
            'quiet': True,
            'no_warnings': True,
            'outtmpl': f"{name}.mp3",
        }
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([video_url])
    except Exception as e:
        st.error(f"Error en la descarga: {e}")

# --- Eliminar archivos antiguos (si existen) ---
if os.path.exists('Cancion.mp3'):
    os.remove('Cancion.mp3')
if os.path.exists('Cancion.wav'):
    os.remove('Cancion.wav')

# --- Entrada del link de la canci√≥n ---
link = st.text_input("Ingrese el link de su canci√≥n de YouTube:")

if st.button("Procesar Canci√≥n"):
    if not link.strip():
        st.error("Por favor, ingrese un link v√°lido.")
    else:
        with st.spinner("Descargando y procesando la canci√≥n..."):
            # Descargar la canci√≥n
            descarga(link, 'Cancion')

            # Verificar si el archivo MP3 fue creado
            if not os.path.exists("Cancion.mp3"):
                st.error("La descarga del archivo MP3 fall√≥.")
            else:
                # Convertir de MP3 a WAV usando ffmpeg (aseg√∫rate de tener ffmpeg instalado)
                try:
                    result = subprocess.run(
                        ['ffmpeg', '-y', '-i', 'Cancion.mp3', 'Cancion.wav'],
                        stdout=subprocess.PIPE,
                        stderr=subprocess.PIPE,
                        text=True
                    )
                    if result.returncode != 0:
                        st.error(f"Error al convertir a WAV: {result.stderr}")
                        st.stop()
                except Exception as e:
                    st.error(f"Error al convertir a WAV: {e}")
                    st.stop()

                st.success("Descarga y conversi√≥n completadas.")

                # --- Procesamiento de la se√±al descargada ---
                # Par√°metros de la muestra
                inicio_seg = 20  # Punto de inicio (en segundos) para extraer la muestra
                tm = 5           # Duraci√≥n de la muestra (en segundos)

                # Leer el archivo WAV
                try:
                    x, fs = sf.read('Cancion.wav')
                except Exception as e:
                    st.error(f"Error al leer el archivo WAV: {e}")
                    st.stop()

                # Verificar que el archivo tenga suficiente duraci√≥n
                if len(x) < int(fs * (inicio_seg + tm)):
                    st.error("El archivo de audio es demasiado corto para extraer la muestra solicitada.")
                    st.stop()

                # Vector de tiempo para la muestra
                t = np.arange(0, tm, 1/fs)

                # Extraer una muestra de la canci√≥n (diferenciando audio mono y est√©reo)
                if x.ndim == 1:
                    # Audio mono
                    M = x[int(fs * inicio_seg):int(fs * (inicio_seg + tm))]
                else:
                    # Audio est√©reo (u otro n√∫mero de canales): se usa el primer canal
                    M = x[int(fs * inicio_seg):int(fs * (inicio_seg + tm)), 0]

                # Se√±al portadora
                Ac = float(max(np.abs(M)))   # Amplitud m√°xima de M
                Fc = 2400 / np.pi            # Frecuencia portadora definida
                C = Ac * np.cos(2 * np.pi * Fc * t)

                # Modulaci√≥n de la se√±al
                Y = (1 + M / Ac) * C

                # --- Figura 1: Se√±al modulada en el tiempo ---
                fig1, ax1 = plt.subplots()
                ax1.plot(t, Y, label="Se√±al modulada", color='purple')
                ax1.set_xlabel('t (s)')
                ax1.set_ylabel('Amplitud')
                ax1.set_title('Se√±al Modulada en el Tiempo')
                ax1.grid(True)
                ax1.legend()
                st.pyplot(fig1)

                # --- Transformadas de Fourier ---
                VF = np.fft.rfftfreq(len(C), 1/fs)  # Vector de frecuencias
                TE = np.fft.rfft(M)                 # Transformada de la se√±al mensaje
                TP = np.fft.rfft(C)                 # Transformada de la portadora
                TM_fft = np.fft.rfft(Y)             # Transformada de la se√±al modulada

                # --- Figura 2: Se√±ales en el dominio del tiempo ---
                fig2, axs2 = plt.subplots(1, 3, figsize=(16, 5))
                axs2[0].plot(t, M, label='Mensaje')
                axs2[0].set_title('Mensaje')
                axs2[0].set_xlabel('t (s)')
                axs2[0].set_ylabel('Amplitud')
                axs2[0].legend()
                axs2[0].grid(True)

                axs2[1].plot(t, C, label='Portadora', color='orange')
                axs2[1].set_title('Portadora')
                axs2[1].set_xlabel('t (s)')
                axs2[1].set_ylabel('Amplitud')
                axs2[1].legend()
                axs2[1].grid(True)

                axs2[2].plot(t, Y, label='Modulaci√≥n', color='green')
                axs2[2].set_title('Modulaci√≥n')
                axs2[2].set_xlabel('t (s)')
                axs2[2].set_ylabel('Amplitud')
                axs2[2].legend()
                axs2[2].grid(True)
                st.pyplot(fig2)

                # --- Figura 3: Se√±ales en el dominio de la frecuencia ---
                fig3, ax3 = plt.subplots(figsize=(16, 5))
                ax3.plot(VF, np.abs(TE), label='Mensaje')
                ax3.plot(VF, np.abs(TP), label='Portadora', linewidth=2)
                ax3.plot(VF, np.abs(TM_fft), label='Modulada', linewidth=1.5)
                ax3.set_title('Amplitud en funci√≥n de la frecuencia')
                ax3.set_xlabel('f (Hz)')
                ax3.set_ylabel('Amplitud')
                ax3.legend()
                ax3.grid(True)
                st.pyplot(fig3)

                # --- Reproducci√≥n de audio ---
                st.markdown("### Reproducci√≥n de Audio")

                # Funci√≥n auxiliar para reproducir audio a partir de un array numpy
                def audio_player(signal, fs, label):
                    buffer = io.BytesIO()
                    sf.write(buffer, signal, fs, format='WAV')
                    buffer.seek(0)
                    st.write(label)
                    st.audio(buffer, format='audio/wav')

                audio_player(M, fs, "Mensaje")
                audio_player(C, fs, "Portadora")
                audio_player(Y, fs, "Se√±al Modulada")
```

**Distorsi√≥n Total de Arm√≥nicos (THD)**:

El THD es una medida de la distorsi√≥n arm√≥nica presente en una se√±al y se define como la relaci√≥n entre la suma de las potencias de todos los componentes arm√≥nicos y la potencia del componente fundamental. Para una se√±al de corriente o tensi√≥n, el THD se calcula como:
$$THD = \frac{\sqrt{V_2^2+V_3^2+\cdots V_n^2}}{V_1}$$


donde Vn es la amplitud del n-√©simo arm√≥nico y V1 es la amplitud del componente fundamental.
El THD indica qu√© tan alejada est√° una se√±al de una senoidal pura. Un THD alto significa m√°s distorsi√≥n.


**C√°lculo del THD usando la FFT**

Podemos calcular el THD aplicando la Transformada R√°pida de Fourier (FFT) a la se√±al de salida del rectificador. Los pasos b√°sicos son:

1.   Aplica la Transformada R√°pida de Fourier (FFT) a la se√±al para obtener su
espectro de frecuencia.
2.   Identifica la amplitud de la frecuencia fundamental (V1) y las amplitudes de los arm√≥nicos (V1,V2,V3,...).


3. Usa la f√≥rmula del THD para calcular la distorsi√≥n.


**Factor de Potencia**:

El factor de potencia (PF) es la relaci√≥n entre la potencia activa (P) y la potencia aparente (S) en un circuito. Se define como:
$$PF_{THD}=\sqrt{\frac{1}{1+(THD)^2}}$$


En presencia de arm√≥nicos, el factor de potencia puede verse afectado por la distorsi√≥n arm√≥nica, lo que lleva a un factor de potencia de desplazamiento (DPF) y un factor de potencia de distorsi√≥n (DF). Un THD alto reduce el FP, lo que afecta la eficiencia del sistema


```python
%%writefile 3_üìä_Distorsion_Total_de_Armonicos.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="THD", page_icon="üìä")

st.markdown("# Punto 3")
st.sidebar.header("Distorsi√≥n Total de Arm√≥nicos")

# Contenido explicativo con ecuaciones en LaTeX
st.markdown(r"""
**Distorsi√≥n Total de Arm√≥nicos (THD):**

El THD es una medida de la distorsi√≥n arm√≥nica presente en una se√±al y se define como la relaci√≥n entre la suma de las potencias de todos los componentes arm√≥nicos y la potencia del componente fundamental. Para una se√±al de corriente o tensi√≥n, el THD se calcula como:

$$
THD = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

donde \(V_n\) es la amplitud del n-√©simo arm√≥nico y \(V_1\) es la amplitud del componente fundamental. El THD indica qu√© tan alejada est√° una se√±al de una senoidal pura. Un THD alto significa mayor distorsi√≥n.

**C√°lculo del THD usando la FFT**

Podemos calcular el THD aplicando la Transformada R√°pida de Fourier (FFT) a la se√±al de salida del rectificador. Los pasos b√°sicos son:

1. Aplica la FFT a la se√±al para obtener su espectro de frecuencia.
2. Identifica la amplitud de la frecuencia fundamental (\(V_1\)) y las amplitudes de los arm√≥nicos (\(V_2, V_3, \dots\)).
3. Usa la f√≥rmula del THD para calcular la distorsi√≥n.

**Factor de Potencia**:

El factor de potencia (PF) es la relaci√≥n entre la potencia activa (P) y la potencia aparente (S) en un circuito. Se define como:

$$
PF_{THD} = \sqrt{\frac{1}{1 + THD^2}}
$$

En presencia de arm√≥nicos, el factor de potencia puede verse afectado por la distorsi√≥n arm√≥nica, lo que lleva a un factor de potencia de desplazamiento (DPF) y un factor de potencia de distorsi√≥n (DF). Un THD alto reduce el FP, lo que afecta la eficiencia del sistema.
""")

# -----------------------------
# SIMULACI√ìN DE LA SE√ëAL Y C√ÅLCULO DEL THD
# -----------------------------

# Par√°metros de la se√±al
frecuencia_fundamental = 50      # Hz
amplitud_fundamental = 10.0       # V
muestras_por_ciclo = 1000
num_ciclos = 5

# Vector de tiempo para los 5 ciclos
tiempo = np.linspace(0, num_ciclos / frecuencia_fundamental,
                     num_ciclos * muestras_por_ciclo, endpoint=False)

# Se√±al fundamental (senoidal)
senal_fundamental = amplitud_fundamental * np.sin(2 * np.pi * frecuencia_fundamental * tiempo)

# Rectificador de onda completa (valor absoluto)
senal_rectificada = np.abs(senal_fundamental)

# Funci√≥n para calcular el THD usando la FFT
def calcular_thd(senal, frecuencia_fundamental, muestras_por_ciclo):
    fft_resultado = np.fft.fft(senal)
    # Se obtiene la mitad del espectro (parte positiva)
    mitad = len(fft_resultado) // 2
    # Se busca el √≠ndice de la frecuencia fundamental en el espectro
    idx_fundamental = np.argmax(np.abs(fft_resultado[:mitad]))
    amplitud_fundamental = np.abs(fft_resultado[idx_fundamental])
    # Suma de las amplitudes de todos los componentes (incluyendo la fundamental)
    suma_total = np.sum(np.abs(fft_resultado[:mitad])**2)
    # THD: se sustrae la contribuci√≥n del componente fundamental
    thd = np.sqrt(suma_total - amplitud_fundamental**2) / amplitud_fundamental
    return thd

# C√°lculo del THD para la se√±al rectificada
thd_rectificada = calcular_thd(senal_rectificada, frecuencia_fundamental, muestras_por_ciclo)
st.write(f"**THD de la se√±al rectificada:** {thd_rectificada:.4f}")

# -----------------------------
# GR√ÅFICOS DE LA SE√ëAL RECTIFICADA Y SU ESPECTRO
# -----------------------------
# Se calcula la FFT para la gr√°fica del espectro
fft_resultado = np.fft.fft(senal_rectificada)
frecuencias = np.fft.fftfreq(len(senal_rectificada), d=1/(frecuencia_fundamental * muestras_por_ciclo))
mitad = len(frecuencias) // 2

fig1, ax1 = plt.subplots(2, 1, figsize=(12, 6))

# Gr√°fica en el dominio del tiempo
ax1[0].plot(tiempo, senal_rectificada, color='C0')
ax1[0].set_title('Se√±al Rectificada de Onda Completa')
ax1[0].set_xlabel('Tiempo [s]')
ax1[0].set_ylabel('Amplitud [V]')
ax1[0].grid(True)

# Gr√°fica en el dominio de la frecuencia (solo la parte positiva)
ax1[1].plot(frecuencias[:mitad], np.abs(fft_resultado[:mitad]), color='C1')
ax1[1].set_title('Espectro de Frecuencia de la Se√±al Rectificada')
ax1[1].set_xlabel('Frecuencia [Hz]')
ax1[1].set_ylabel('Amplitud')
ax1[1].set_xlim(0, 500)
ax1[1].grid(True)

fig1.tight_layout()
st.pyplot(fig1)

# -----------------------------
# SIMULACI√ìN DE UNA CARGA RC
# -----------------------------
R = 100        # Ohm
C = 100e-6     # F
tau = R * C

# Simulaci√≥n de la respuesta RC usando integraci√≥n num√©rica
senal_rc = np.zeros_like(tiempo)
for i in range(1, len(tiempo)):
    dt = tiempo[i] - tiempo[i-1]
    senal_rc[i] = senal_rc[i-1] + (senal_rectificada[i] - senal_rc[i-1]) * dt / tau

# C√°lculo del THD para la se√±al con carga RC
thd_rc = calcular_thd(senal_rc, frecuencia_fundamental, muestras_por_ciclo)
st.write(f"**THD de la se√±al con carga RC:** {thd_rc:.4f}")

# Gr√°ficos de la respuesta RC y su espectro
fft_resultado_rc = np.fft.fft(senal_rc)
frecuencias_rc = np.fft.fftfreq(len(senal_rc), d=1/(frecuencia_fundamental * muestras_por_ciclo))
mitad_rc = len(frecuencias_rc) // 2

fig2, ax2 = plt.subplots(2, 1, figsize=(12, 6))

# Se√±al con carga RC en el dominio del tiempo
ax2[0].plot(tiempo, senal_rc, color='C2')
ax2[0].set_title('Se√±al Rectificada con Carga RC')
ax2[0].set_xlabel('Tiempo [s]')
ax2[0].set_ylabel('Amplitud [V]')
ax2[0].grid(True)

# Espectro de la se√±al con carga RC
ax2[1].plot(frecuencias_rc[:mitad_rc], np.abs(fft_resultado_rc[:mitad_rc]), color='C3')
ax2[1].set_title('Espectro de Frecuencia de la Se√±al con Carga RC')
ax2[1].set_xlabel('Frecuencia [Hz]')
ax2[1].set_ylabel('Amplitud')
ax2[1].grid(True)

fig2.tight_layout()
st.pyplot(fig2)

```

```python
%%writefile 4_üéôÔ∏è_Detector_de_Voz.py
import streamlit as st
import pandas as pd
import altair as alt
import numpy as np
import matplotlib.pyplot as plt
import librosa
# from pydub import AudioSegment  # Eliminado porque no se utiliza
import os
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import cosine
import soundfile as sf
import io
import gdown

# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="Detector de Voz", page_icon="üéôÔ∏è")

st.markdown("# Punto 4")
st.sidebar.header("Detector de Voz")
st.write(
    """Esta demo muestra c√≥mo visualizar y procesar audios para la detecci√≥n de voz.
Se descargan audios de referencia desde Google Drive, se procesan (se eliminan silencios, se calcula la FFT, se normaliza el espectro) y se comparan para verificar el acceso.
"""
)

# Diccionario con los ID de cada archivo de referencia (reemplaza 'ID_REAL_X' por los IDs correctos)
file_ids = {
    "clave1": "1WniR2Rd1KnoIAHhGmwh9k5WJoaepU4sJ",  # base de datos
    "clave2": "1J9jsg4G3duHX9OOwJxLuNuP016wjIc9b"
}

# Funci√≥n para descargar un archivo desde Google Drive usando gdown
def descargar_audio(file_id, output_filename):
    url = f"https://drive.google.com/uc?id={file_id}"
    st.write(f"Descargando {output_filename} desde {url}...")
    gdown.download(url, output_filename, quiet=False)

# Descargamos los audios de referencia (si no existen ya)
for key, file_id in file_ids.items():
    output_file = f"{key}.wav"
    if not os.path.exists(output_file):
        descargar_audio(file_id, output_file)
    else:
        st.write(f"Archivo {output_file} ya existe.")

# --- Visualizaci√≥n de cada audio en el dominio del tiempo ---
st.markdown("### Visualizaci√≥n de audios (dominio del tiempo)")
for key in file_ids.keys():
    filename = f"{key}.wav"
    audio, sr = librosa.load(filename, sr=None)
    st.write(f"**Audio {key}:** Duraci√≥n = {len(audio)/sr:.2f} s, Tasa de muestreo = {sr} Hz")
    tiempo = np.linspace(0, len(audio)/sr, num=len(audio))

    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(tiempo, audio, color='C0')
    ax.set_xlabel("Tiempo (s)")
    ax.set_ylabel("Amplitud")
    ax.set_title(f"Se√±al de audio para {key}")
    ax.grid(True)
    st.pyplot(fig)

# --- Transformada de Fourier de cada audio ---
st.markdown("### Espectro de Fourier de cada audio")
for key in file_ids.keys():
    filename = f"{key}.wav"
    audio, sr = librosa.load(filename, sr=None)
    fft_audio = np.fft.fft(audio)
    fft_audio = np.abs(fft_audio[:len(fft_audio)//2])
    freqs = np.linspace(0, sr/2, num=len(fft_audio))

    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(freqs, fft_audio, color='C1')
    ax.set_xlabel("Frecuencia (Hz)")
    ax.set_ylabel("Magnitud")
    ax.set_title(f"Espectro de Fourier para {key}")
    ax.grid(True)
    st.pyplot(fig)

# --- Procesamiento: Eliminaci√≥n de silencios ---
st.markdown("### Eliminaci√≥n de silencios en los audios")
signals_no_silence = {}

for key in file_ids.keys():
    filename = f"{key}.wav"
    audio, sr = librosa.load(filename, sr=None)
    intervals = librosa.effects.split(audio, top_db=35)
    audio_no_silence = np.concatenate([audio[start:end] for start, end in intervals])

    signals_no_silence[key] = {
        "signal": audio_no_silence,
        "sampling_rate": sr
    }

    tiempo_no_silence = np.linspace(0, len(audio_no_silence) / sr, num=len(audio_no_silence))

    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(tiempo_no_silence, audio_no_silence, color='C2')
    ax.set_xlabel("Tiempo (s)")
    ax.set_ylabel("Amplitud")
    ax.set_title(f"Se√±al sin silencio para {key}")
    ax.grid(True)
    st.pyplot(fig)

st.write("Se√±ales sin silencio almacenadas:", list(signals_no_silence.keys()))

# --- Reproducci√≥n de las se√±ales sin silencios ---
st.markdown("### Reproducci√≥n de audios (sin silencios)")
def reproducir_audio(signal, sr, label):
    # Escribir el audio a un buffer en formato WAV
    buffer = io.BytesIO()
    sf.write(buffer, signal, sr, format='WAV')
    buffer.seek(0)
    st.write(label)
    st.audio(buffer, format='audio/wav')

for key in file_ids.keys():
    data = signals_no_silence[key]
    reproducir_audio(data["signal"], data["sampling_rate"], f"Audio sin silencios: {key}")

# --- C√°lculo y visualizaci√≥n de la FFT de las se√±ales sin silencios ---
st.markdown("### FFT de las se√±ales sin silencios")
fourier_signals = {}

for key, data in signals_no_silence.items():
    signal = data["signal"]
    sr = data["sampling_rate"]
    fft_signal = np.fft.fft(signal)
    freqs = np.fft.fftfreq(len(signal), d=1/sr)
    fourier_signals[key] = {
        "fft": fft_signal,
        "frequencies": freqs
    }
    idx_positive = np.where(freqs >= 0)
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(freqs[idx_positive], np.abs(fft_signal)[idx_positive], color='C3')
    ax.set_xlabel("Frecuencia (Hz)")
    ax.set_ylabel("Magnitud")
    ax.set_title(f"Espectro (FFT) de la se√±al para {key}")
    ax.grid(True)
    st.pyplot(fig)

# --- Normalizaci√≥n del espectro ---
st.markdown("### Espectro normalizado")
fourier_signals_normalized = {}

for key, data in fourier_signals.items():
    fft_signal = data["fft"]
    frequencies = data["frequencies"]
    idx_positive = frequencies >= 0
    vf = frequencies[idx_positive]
    Xw = np.abs(fft_signal[idx_positive])
    sca = MinMaxScaler()
    Xw_norm = sca.fit_transform(Xw.reshape(-1, 1)).flatten()

    fourier_signals_normalized[key] = {
        "normalized_spectrum": Xw_norm,
        "frequencies": vf
    }

    # Gr√°fica del espectro normalizado
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(vf, Xw_norm, color='C4')
    ax.set_xlabel('f [Hz]')
    ax.set_ylabel('|X(f)|')
    ax.set_title(f"Espectro normalizado para {key}")
    ax.grid(True)
    st.pyplot(fig)

    # Gr√°fica en dB
    fig_db, ax_db = plt.subplots(figsize=(14, 4))
    ax_db.plot(vf, 20 * np.log10(Xw_norm + 1e-10), color='C5')
    ax_db.set_xlabel('f [Hz]')
    ax_db.set_ylabel('|X(f)| dB')
    ax_db.set_title(f"Espectro normalizado en dB para {key}")
    ax_db.grid(True)
    st.pyplot(fig_db)

# --- Funci√≥n para obtener espectro normalizado ---
def obtener_espectro_normalizado(filename, top_db=35):
    audio, sr = librosa.load(filename, sr=None)
    intervals = librosa.effects.split(audio, top_db=top_db)
    audio_no_silence = np.concatenate([audio[start:end] for start, end in intervals])
    fft_audio = np.fft.fft(audio_no_silence)
    fft_audio = np.abs(fft_audio[:len(fft_audio)//2])
    scaler = MinMaxScaler()
    fft_audio_norm = scaler.fit_transform(fft_audio.reshape(-1, 1)).flatten()
    return fft_audio_norm, sr

# Procesar audios de referencia
spec_ref1, sr1 = obtener_espectro_normalizado("clave1.wav")
spec_ref2, sr2 = obtener_espectro_normalizado("clave2.wav")

min_len_refs = min(len(spec_ref1), len(spec_ref2))
spec_ref1 = spec_ref1[:min_len_refs]
spec_ref2 = spec_ref2[:min_len_refs]

# Calcular la distancia de coseno entre las referencias
d_ref = cosine(spec_ref1, spec_ref2)
st.write(f"**Distancia de referencia (clave1 vs clave2):** {d_ref:.4f}")

tolerancia = 0.05
umbral_min = d_ref - tolerancia
umbral_max = d_ref + tolerancia
st.write(f"**Umbral de aceptaci√≥n:** [{umbral_min:.4f}, {umbral_max:.4f}]")

# --- Funci√≥n para verificar acceso con un nuevo audio ---
def verificar_acceso(nuevo_audio_filename, ref1, ref2, umbral_min, umbral_max, top_db=35):
    spec_nuevo, sr_nuevo = obtener_espectro_normalizado(nuevo_audio_filename, top_db=top_db)
    min_len = min(len(spec_nuevo), len(ref1), len(ref2))
    spec_nuevo = spec_nuevo[:min_len]
    ref1_cut = ref1[:min_len]
    ref2_cut = ref2[:min_len]
    d1 = cosine(spec_nuevo, ref1_cut)
    d2 = cosine(spec_nuevo, ref2_cut)
    st.write(f"Distancia con clave1: {d1:.4f}")
    st.write(f"Distancia con clave2: {d2:.4f}")
    if umbral_min <= d1 <= umbral_max or umbral_min <= d2 <= umbral_max:
        st.write("**Acceso autorizado.**")
    else:
        st.write("**Acceso denegado.**")

# --- Carga de nuevo audio para verificaci√≥n ---
st.markdown("### Verificar acceso con nueva muestra de voz")
uploaded_file = st.file_uploader("Sube un archivo de audio (wav o mp3) que contenga la palabra clave", type=["wav", "mp3"])

if uploaded_file is not None:
    # Guardar el archivo subido como "nuevo.wav"
    nuevo_audio_filename = "nuevo.wav"
    with open(nuevo_audio_filename, "wb") as f:
        f.write(uploaded_file.getbuffer())
    st.success("Archivo subido correctamente.")
    # Reproducir el audio subido
    audio_data, sr_nuevo = librosa.load(nuevo_audio_filename, sr=None)
    reproducir_audio(audio_data, sr_nuevo, "Audio subido (nuevo)")
    # Verificar acceso
    verificar_acceso(nuevo_audio_filename, spec_ref1, spec_ref2, umbral_min, umbral_max)

# --- Simulaci√≥n de extracci√≥n de caracter√≠sticas y t-SNE ---
st.markdown("### An√°lisis de caracter√≠sticas con t-SNE")
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Generamos datos sint√©ticos para representar caracter√≠sticas (por ejemplo, MFCCs)
np.random.seed(42)
clave1_features = np.random.normal(loc=0, scale=1, size=(50, 13))
clave2_features = np.random.normal(loc=0.2, scale=1, size=(50, 13))
nuevo_audio_features = np.random.normal(loc=3, scale=1, size=(50, 13))

X = np.vstack([clave1_features, clave2_features, nuevo_audio_features])
y = np.array([1]*50 + [1]*50 + [2]*50)  # 1 para clave1 y clave2; 2 para nuevo audio

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(X_scaled)

fig_tsne, ax_tsne = plt.subplots(figsize=(8, 6))
scatter = ax_tsne.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap="viridis", alpha=0.7)
ax_tsne.set_title("Separabilidad de los Audios con t-SNE")
ax_tsne.set_xlabel("Componente 1")
ax_tsne.set_ylabel("Componente 2")
ax_tsne.grid(True)
st.pyplot(fig_tsne)
```
